```python
# Import required libraries
import pandas as pd

# ---------------------------------------------------
# 1. Load the CSV file and display the first 5 rows
# ---------------------------------------------------

# Load the dataset (make sure the file is uploaded to Colab)
df = pd.read_csv("stress_strain_practice.csv")

# Display first 5 rows
print("First 5 rows of the dataset:")
display(df.head())

# ---------------------------------------------------
# 2. Calculate Stress (MPa) = Force (N) / Area (mm^2)
#    Note: 1 N/mm^2 = 1 MPa
# ---------------------------------------------------

df["Stress_MPa"] = df["Force_N"] / df["Cross_Section_Area_mm2"]

# ---------------------------------------------------
# 3. Calculate Strain (%) = (Elongation / Original Length) * 100
# ---------------------------------------------------

df["Strain_Percent"] = (df["Elongation_mm"] / df["Original_Length_mm"]) * 100

# ---------------------------------------------------
# 4. Group by Material and calculate average Stress & Strain
# ---------------------------------------------------

material_avg = df.groupby("Material")[["Stress_MPa", "Strain_Percent"]].mean()

print("\nAverage Stress and Strain by Material:")
display(material_avg)

# ---------------------------------------------------
# 5. Identify the strongest material (highest average stress)
# ---------------------------------------------------

strongest_material = material_avg["Stress_MPa"].idxmax()
max_stress_value = material_avg["Stress_MPa"].max()

print(f"\nStrongest Material: {strongest_material}")
print(f"Highest Average Stress: {max_stress_value:.2f} MPa")

# ---------------------------------------------------
# 6. Summary: Min, Max, and Average Stress for each Material
# ---------------------------------------------------

stress_summary = df.groupby("Material")["Stress_MPa"].agg(
    Min_Stress="min",
    Max_Stress="max",
    Avg_Stress="mean"
)

print("\nStress Summary by Material:")
display(stress_summary)

# ---------------------------------------------------
# 7. Export results to a new CSV file
# ---------------------------------------------------

# Combine useful results into one DataFrame for export
final_output = df.merge(
    material_avg,
    on="Material",
    suffixes=("", "_Material_Avg")
)

# Save to CSV
final_output.to_csv("stress_strain_analysis_results.csv", index=False)

print("\nResults exported successfully to 'stress_strain_analysis_results.csv'")
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------------------
# 1. Load the CSV file with basic error handling
# ---------------------------------------------------

try:
    df = pd.read_csv("stress_strain_practice.csv")
    print("File loaded successfully.")
except FileNotFoundError:
    print("Error: File not found. Please upload the CSV file.")
    raise

# Display first 5 rows
display(df.head())

# ---------------------------------------------------
# 2. Handle missing values
# ---------------------------------------------------
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Drop rows with missing critical values
df = df.dropna(subset=[
    "Force_N",
    "Cross_Section_Area_mm2",
    "Elongation_mm",
    "Original_Length_mm",
    "Material"
])

print("\nRows after removing missing values:", len(df))

# ---------------------------------------------------
# 3. Calculate Stress (MPa) and Strain (%)
# ---------------------------------------------------

df["Stress_MPa"] = df["Force_N"] / df["Cross_Section_Area_mm2"]
df["Strain_Percent"] = (df["Elongation_mm"] / df["Original_Length_mm"]) * 100

# ---------------------------------------------------
# 4. Highlight tests where Stress > 200 MPa
# ---------------------------------------------------

high_stress_tests = df[df["Stress_MPa"] > 200]

print("\nâš ï¸ Tests where Stress exceeded 200 MPa:")
display(high_stress_tests)

# Optional: Add a flag column
df["High_Stress_Flag"] = df["Stress_MPa"] > 200

# ---------------------------------------------------
# 5. Group by Material and calculate average Stress & Strain
# ---------------------------------------------------

material_avg = df.groupby("Material")[["Stress_MPa", "Strain_Percent"]].mean()

display(material_avg)

# ---------------------------------------------------
# 6. Identify the strongest material
# ---------------------------------------------------

strongest_material = material_avg["Stress_MPa"].idxmax()
print(f"\nStrongest Material based on average stress: {strongest_material}")

# ---------------------------------------------------
# 7. Visualization: Bar chart of Average Stress by Material
# ---------------------------------------------------

plt.figure()
material_avg["Stress_MPa"].plot(kind="bar")
plt.title("Average Stress by Material")
plt.xlabel("Material")
plt.ylabel("Average Stress (MPa)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 8. Stress Summary (Min, Max, Average)
# ---------------------------------------------------

stress_summary = df.groupby("Material")["Stress_MPa"].agg(
    Min_Stress="min",
    Max_Stress="max",
    Avg_Stress="mean"
)

display(stress_summary)

# ---------------------------------------------------
# 9. Export final results to CSV
# ---------------------------------------------------

final_output = df.merge(
    stress_summary,
    on="Material"
)

final_output.to_csv("stress_strain_analysis_results.csv", index=False)

print("\nAnalysis complete. Results saved to 'stress_strain_analysis_results.csv'")
```
